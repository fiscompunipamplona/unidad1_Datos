```python
import numpy as np
from scipy.stats import cauchy, t
import matplotlib.pyplot as plt
```

```python
# Normal distribution

mu = 0
sg = 1
nevents = 1e4

norm_d = np.random.normal(mu, sg, int(nevents) )

plt.plot(norm_d)
```

```python
st_bin = -5
ed_bin = 5
nb = 20
bw = (ed_bin - st_bin)/nb

norm_dist = np.zeros(nb)

for i in range(int(nevents)):
    tmp = norm_d[i]/bw
    norm_dist[ int( 10.5 + tmp ) ] += 1
        
x = []
[ x.append(i) for i in np.arange(st_bin,ed_bin, bw)]
plt.plot(x,norm_dist,'o')
plt.grid('on',ls='-.')
plt.xlim(-4.5, 4.5)
```

```python
# Cauchy distribution

mu = 0
sg = 1
nevents = 1e4

x_c = np.linspace(cauchy.ppf(0.1),cauchy.ppf(.9), 20)

plt.plot(x_c, cauchy.pdf(x_c),'ro') #, lw=1, alpha=0.6, label='cauchy pdf')
```
